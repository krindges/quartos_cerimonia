```python
import random
import pandas as pd
endereco1=r'C:\Users\USER\Dropbox\Projetos_python\quartos_cerimonia\Estudo_roomlist2025.xlsx'
Q = pd.read_excel(endereco1, sheet_name='Planilha1', index_col=None)#, skiprows=1
meninos=Q[Q['Sexo']=='Masculino'].reset_index()
meninas=Q[Q['Sexo']=='Feminino'].reset_index()
meninas.loc[:, 'Nasc'] = pd.to_datetime(meninas['Nasc'], format="%d/%m/%Y")
meninos.loc[:, 'Nasc'] = pd.to_datetime(meninos['Nasc'], format="%d/%m/%Y")
meninos['Nasc'] = pd.to_datetime(meninos['Nasc'], errors='coerce', dayfirst=True)
meninas['Nasc'] = pd.to_datetime(meninas['Nasc'], errors='coerce', dayfirst=True)
meninos['Dias desde nascimento'] = (pd.Timestamp.today() - meninos['Nasc']).dt.days
meninas['Dias desde nascimento'] = (pd.Timestamp.today() - meninas['Nasc']).dt.days
meninos = meninos.sort_values(by='Dias desde nascimento', ascending=False)
meninas = meninas.sort_values(by='Dias desde nascimento', ascending=False)

def alocar_criancas_em_quartos(Q):
    # Número total de crianças
    num_criancas = len(Q)

    # Determinar o número de quartos (cada quarto tem no máximo 3 crianças)
    num_quartos = math.ceil(num_criancas / 3)

    # Criar vetor de capacidades de cada quarto
    capacidade_quarto = 3 * np.ones((num_quartos, 1), dtype=int)

    # Ajustar o último quarto caso sobre menos de 3 crianças
    resto = num_criancas % 3
    if resto > 0:
        capacidade_quarto[-1][0] = resto
    
    # Gerar idades aleatórias entre 6 e 15 anos
    idades = Q['Dias desde nascimento'].to_list()
    
    # Ordenar crianças por idade
    criancas_ordenadas = sorted(enumerate(idades), key=lambda x: x[1])
    
    # Criar alocação de quartos
    alocacao = {}
    for j in range(num_quartos):
        alocacao[j] = []
    
    # Distribuir crianças em grupos de 3 minimizando a diferença de idade
    for i in range(num_quartos):
        grupo = criancas_ordenadas[i * capacidade_quarto[i][0]:(i + 1) * capacidade_quarto[i][0]]
        alocacao[i] = [crianca[0] for crianca in grupo]
    
    return alocacao, idades

# Executar a função
alocacao, idades = alocar_criancas_em_quartos(meninos)
if alocacao:
    for quarto, criancas in alocacao.items():
        print(f"Quarto {quarto}: Crianças {criancas} com idades {[idades[i] for i in criancas]}")
```

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
def distancias(df):
    num_cidades = len(df)
    df['Cidade_Estado'] = df['Cidade'] + ' - ' + df['Estado']
    matriz_distancias = np.zeros((num_cidades, num_cidades))

    # Calcular a matriz de distâncias
    for i in range(num_cidades):
        for j in range(num_cidades):
            if i != j:
                cidade1 = (df.loc[i, "Latitude"], df.loc[i, "Longitude"])
                cidade2 = (df.loc[j, "Latitude"], df.loc[j, "Longitude"])
                matriz_distancias[i, j] = geodesic(cidade1, cidade2).kilometers

    # Converter a matriz para um DataFrame pandas
    df_distancias = pd.DataFrame(matriz_distancias, index=df['Cidade_Estado'], columns=df['Cidade_Estado'])

    # Exibir a matriz de distâncias
    print(df_distancias)

    # Salvar a matriz de distâncias em um arquivo Excel
    df_distancias.to_excel("matriz_distancias.xlsx")
    return df_distancias
ssss=distancias(meninas)
```
